```python
df_daily = df_copy.resample(rule='D').sum()
```

```python
df_daily = df_daily.replace(0.0, np.nan)
df_daily.dropna(inplace=True)
```

```python
from scipy.stats import boxcox
df_daily['y'], lam = boxcox(df_daily['y'])
print('Lambda: %f' % lam)
```

```python
df_daily.plot()
```

```python
df_daily.reset_index(inplace=True)
```

```python
m_daily = Prophet()
```

```python
m_daily.fit(df_daily)
```

```python
future = m_daily.make_future_dataframe(periods=365)
future.tail()
```

```python
forecast = m_daily.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
fig1 = m_daily.plot(forecast)
```

```python
fig4 = m_daily.plot_components(forecast)
```

```python
m_daily_sea = Prophet(daily_seasonality=True)

m_daily_sea.fit(df_daily)

future = m_daily_sea.make_future_dataframe(periods=365)
future.tail()

forecast = m_daily_sea.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()

fig2 = m_daily_sea.plot(forecast)
```

```python
fig5 = m_daily_sea.plot_components(forecast)
```

```python
df_monthly = df_copy.resample(rule='M').sum()

from scipy.stats import boxcox
df_monthly['y'], lam = boxcox(df_monthly['y'])
print('Lambda: %f' % lam)


df_monthly.plot()
```

```python
df_monthly.reset_index(inplace=True)
```

```python
m_monthly = Prophet()
```

```python
future = m_monthly.make_future_dataframe(periods=365)
future.tail()
```
